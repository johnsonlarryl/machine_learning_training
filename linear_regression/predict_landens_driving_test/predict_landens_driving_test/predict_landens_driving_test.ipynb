```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
```

```python
driving_test_results = pd.read_csv("../test/data/landen_driving_history.csv")
```

```python
driving_test_results.shape
```

```python
driving_test_results
```

```python
x = driving_test_results[['taking_the_test','maintaining_vehicle_position','using_signals','turning','controlling_the_vehicle','approaching_intersections','stopping_in_regular_traffic_conditions','observing_traffic','backing_in_a_straight_line','parallel_parking']]
y = driving_test_results['average_score']
```

```python
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(x, y)
```

```python
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
# with statsmodels
x = sm.add_constant(x) # adding a constant
```

```python
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
```

```python
print_model = model.summary()
print(print_model)
```

```python
taking_the_test = 10
maintaining_vehicle_position = 7.00
using_signals = 8.00
turning = 6.50
controlling_the_vehicle = 6.44
approaching_intersections = 6.75
stopping_in_regular_traffic_conditions = 6.94
observing_traffic = 7.43
backing_in_a_straight_line = 3.17
parallel_parking = 1.00
```

```python
model.params
```

```python
predicted_driving_score = model.params[0] * taking_the_test + \
                          model.params[1] * maintaining_vehicle_position + \
                          model.params[2] * using_signals + \
                          model.params[3] * turning + \
                          model.params[4] * controlling_the_vehicle + \
                          model.params[5] * approaching_intersections + \
                          model.params[6] * stopping_in_regular_traffic_conditions + \
                          model.params[7] * observing_traffic + \
                          model.params[8] * backing_in_a_straight_line + \
                          model.params[8] * parallel_parking
```

```python
print(f"Landen's predicted driving score for assumming driving test is based on a score of 100 is {round(predicted_driving_score, 2)}. ")
```
