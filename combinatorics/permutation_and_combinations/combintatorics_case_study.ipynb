```python
# Cheese, a little league baseball player in the Negro League Classic tournament is trying
# to understand how many times each team will play each other during the regulary season; 
# assumming each time will play each other one time.  If there are 6 teams in the league, how 
# many games will be played in total.
```

```python
# Import dependencies
from ast import literal_eval
import csv
import pandas as pd
from permutation_and_combinations.combinatorics import COMBINATION, combination, combinatorics_list
```

```python
# Read in data
teams_file = csv.DictReader(open("permutation_and_combinations/data/teams.csv"))
```

```python
# Total games played in teams in the Negro League Classic tournament
teams = []
r = 2

for row in teams_file:
    teams.append(row)

team_ids = [team["id"] for team in teams]
```

```python
games = combinatorics_list(team_ids, r, algo=COMBINATION)
print(f"Total number of teams in the {len(games)}")
```

```python
def get_team_name(id: int):
    try:
        return teams[int(id) - 1].get('name')
    except Exception as e:
        print(e)
        return "Unknown Team Name"
```

```python
games_df = pd.DataFrame(games, columns =['Team A', 'Team B'])
```

```python
games_df
```

```python
games_df["Team A"] = games_df["Team A"].apply(get_team_name)
games_df["Team B"] = games_df["Team B"].apply(get_team_name)
```
